```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd

# 設定 data_path
dir_data = '../data/'
```

```python
f_app_train = os.path.join(dir_data, 'application_train.csv')
app_train = pd.read_csv(f_app_train)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

觀察有興趣的欄位的資料分佈，並嘗試找出有趣的訊息
#### Eg

- 畫出任意欄位的[直方圖](https://zh.wikipedia.org/zh-tw/%E7%9B%B4%E6%96%B9%E5%9B%BE)

#### 顯示所有數值欄位的平均數及標準差

```python
for index, col in enumerate(app_train):
    # 判斷是否為數值(整數及實數)
    if app_train[col].dtype == 'int64' or app_train[col].dtype == 'float64':
        # 將數值類數小於 2 的欄位判定為可能的類別欄位，只去計算大於 2 個類的欄位其平均數及標準差
        if len(list(app_train[col].unique())) > 2:
            # 數值欄位
            print('第 %3d 欄 平均數: %f \t標準差: %f \t名稱: %s' % (index + 1, app_train[col].mean(), app_train[col].std(), col))
#         else:
#             # 可能的類別欄位
#     else:
#         # 非數值欄位(類別欄位)
```

#### 顯示前十欄為數值欄位的統計數據

```python
for index, col in enumerate(app_train):
    # 判斷前十欄
    if index < 10:
        # 判斷是否為數值(整數及實數)
        if app_train[col].dtype == 'int64' or app_train[col].dtype == 'float64':
            # 將數值類數小於 2 的欄位判定為可能的類別欄位，只去計算大於 2 個類的欄位其平均數及標準差
            if len(list(app_train[col].unique())) > 2:
                # 數值欄位
                print('第 %2d 欄\n%s\n' % (index + 1, app_train[col].describe()))
#             else:
#                 # 可能的類別欄位
#         else:
#             # 非數值欄位(類別欄位)
```

#### 顯示前十欄可能為類別欄位的數字欄位及所有類別超過十類的類別欄位的直方圖

```python
for index, col in enumerate(app_train):
    # 判斷是否為數值(整數及實數)
    if app_train[col].dtype == 'int64' or app_train[col].dtype == 'float64':
        # 將數值類數小於 2 的欄位判定為可能的類別欄位，只去計算大於 2 個類的欄位其平均數及標準差
        if len(list(app_train[col].unique())) <= 2 and index < 10:
            # 可能的類別欄位
            # 值被數字化可以直接套用 hist()
            print('%s\n' % col)
            app_train[col].hist()
            # 不先執行 show() 的話，會在最後一起被 dump 出來
            plt.show()
#             else:
#                 # 數值欄位
    else:
        # 非數值欄位(類別欄位)
        # 非數字化的值需要經過 value_counts()
        if len(list(app_train[col].unique())) > 10:
            print('%s\n' % col)
            app_train[col].value_counts().plot.bar()
            plt.show()
```
