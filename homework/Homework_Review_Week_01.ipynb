```python
# Day_001 程式碼歸納與補充

import numpy as np
import matplotlib.pyplot as plt

w = 3
b = 0.5

# 生成有序數列
x = np.linspace(0, 100, 101) # 等差數列

# 生成無序數列
y = (x + np.random.randn(101) * 5) * w + b
# np.random.rand(101)  # 非常態數列
# np.random.randn(101) # 常態數列

plt.plot(x, y, 'b.', label = 'data points') # 畫出藍色點圖並定義其名稱
plt.title("Assume we have data points")     # 設定圖片標題
plt.legend(loc = 2)                         # 設定圖示的象限
plt.show()                                  # 輸出圖片

y_hat = x * w + b # 設定預測值
plt.plot(x, y, 'b.', label = 'data')
plt.plot(x, y_hat, 'r-', label = 'prediction')
plt.title("Assume we have data points (And the prediction)")
plt.legend(loc = 2)
plt.show()

# python 函式的定義及數學式 sum abs pow 等等的應用
def mean_absolute_error(y, yp):
    mae = MAE = sum(abs(y - yp)) / len(y)
    return mae

def mean_squared_error(y, yp):
    mse = sum((y - yp) ** 2) / len(y)
#     mse = sum(pow((y - yp), 2)) / len(y) # 另一種寫法
    return mse

MSE = mean_squared_error(y, y_hat)
MAE = mean_absolute_error(y, y_hat)
print("The Mean squared error is %.3f" % (MSE)) # .3f 只輸出小數點後三位
print("The Mean absolute error is %.3f" % (MAE))
```

```python
import requests

# 從網路上讀取檔案
target_url = 'https://raw.githubusercontent.com/vashineyu/slides_and_others/master/tutorial/examples/imagenet_urls_examples.txt'
response = requests.get(target_url) # http 狀態碼
data = response.text                # 指派回傳的內容值並將其字串化

# 用 request 傳送回來的資料不會認得斷行符號
print(len(data), '\n') # 回傳資料的長度
print(data[0:100])     # print 函式會自動將 unicode 符號呈現出來
data[0:100]            # 一般程式區塊後面的 dump 會將 unicode 原本的樣子輸出
```

```python
data = data.split("\n") # 以換行符號切割原本的字元陣列
print(len(data), '\n')
print(data[0])
data[0]
```

```python
# 將 txt 轉成 pandas dataframe
import pandas as pd

arrange_data = []
for d in data:
    line = d.split("\t") # 以 tab 符號來切割字串
    arrange_data.append(line)

df = pd.DataFrame(arrange_data)
print(df.head(), '\n')

# 懶人解法：直接讀取網路連結並分割
df_lazy = pd.read_csv(target_url, sep="\t", header=None) # 需要下載檔案的時間
print(df_lazy.head())
```

```python
from PIL import Image
from io import BytesIO
import numpy as np
import matplotlib.pyplot as plt

response = requests.get(df.loc[0, 1]) # 先讀取第一筆資料的圖片
img = Image.open(BytesIO(response.content))
img = np.array(img)
print(img.shape)
plt.imshow(img)
plt.show()
```

```python
def img2arr_fromURLs(url_list, resize = False):
    img_list = []
    for url in url_list:
        response = requests.get(url)
        try:
            img = Image.open(BytesIO(response.content))
            if resize:
                img = img.resize((256,256)) # 假如 resize, 就先統一到 256 x 256
            img = np.array(img)
            img_list.append(img)
        except:
            # 只有在 response.status_code 為 200 時，才可以取得圖片，若有 404 或其他 status code, 會碰到 Error, 所以我們用 Try 語法避開取不到的狀況
            pass
    
    return img_list

result = img2arr_fromURLs(df[0:5][1].values)
print("Total images that we got: %i " % len(result)) # 如果不等於 5, 代表有些連結失效囉

for im_get in result:
    plt.imshow(im_get)
    plt.show()
```

```python
# Day_006 程式碼歸納與補充

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

# 設定 data_path
dir_data = '../data/'
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

```python
# 先篩選數值型的欄位
# numeric_columns = list(app_train.columns[list(app_train.dtypes.isin(['int64', 'float64']))])
dtype_select = ['float64', 'int64']
numeric_columns = list(app_train.select_dtypes(include=dtype_select))

# 再把只有 2 值 (通常是 0,1) 的欄位去掉
numeric_columns = list(app_train[numeric_columns].columns[list(app_train[numeric_columns].apply(lambda x:len(x.unique())!=2))])
print("Numbers of remain columns: %d" % len(numeric_columns))

# 檢視這些欄位的數值範圍
for col in numeric_columns:
    app_train.boxplot(col)
    plt.show()
```

```python
# 從上面的圖檢查的結果，至少這三個欄位好像有點可疑

# AMT_INCOME_TOTAL
# REGION_POPULATION_RELATIVE
# OBS_60_CNT_SOCIAL_CIRCLE
```

```python
# 最大值離平均與中位數很遠
print(app_train['AMT_INCOME_TOTAL'].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)
cdf = app_train.AMT_INCOME_TOTAL.value_counts().sort_index().cumsum()

plt.plot(list(cdf.index), cdf/cdf.max())
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.xlim([cdf.index.min(), cdf.index.max() * 1.05]) # 限制顯示圖片的範圍
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()

plt.plot(np.log(list(cdf.index)), cdf/cdf.max())
plt.xlabel('Value (log-scale)')
plt.ylabel('ECDF')

plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()
```

```python
# 最大值落在分布之外
print(app_train['REGION_POPULATION_RELATIVE'].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)
cdf = app_train.REGION_POPULATION_RELATIVE.value_counts().sort_index().cumsum()
plt.plot(list(cdf.index), cdf/cdf.max())
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍
plt.show()

app_train['REGION_POPULATION_RELATIVE'].hist()
plt.show()

app_train['REGION_POPULATION_RELATIVE'].value_counts()

# 就以這個欄位來說，雖然有資料掉在分布以外，也不算異常，僅代表這間公司在稍微熱鬧的地區有的據點較少，
# 導致 region population relative 在少的部分較為密集，但在大的部分較為疏漏
```

```python
# 最大值落在分布之外
print(app_train['OBS_60_CNT_SOCIAL_CIRCLE'].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)
cdf = app_train.OBS_60_CNT_SOCIAL_CIRCLE.value_counts().sort_index().cumsum()
plt.plot(list(cdf.index), cdf/cdf.max())
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.xlim([cdf.index.min() * 0.95, cdf.index.max() * 1.05])
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍
plt.show()

app_train['OBS_60_CNT_SOCIAL_CIRCLE'].hist()
plt.show()

app_train['OBS_60_CNT_SOCIAL_CIRCLE'].value_counts().sort_index(ascending = False)
# 注意：當 histogram 畫出上面這種圖 (只出現一條，但是 x 軸延伸很長導致右邊有一大片空白時
# ，代表右邊有值但是數量稀少。這時可以考慮用 value_counts 去找到這些數值
```

```python
# ECDF補充

def ecdf(data):
    x = np.sort(data)
    y = np.arange(1, len(x)+1) / len(x)
    return (x, y)

def plot_ecdf(data, xlabel, ylabel='ECDF'):
    x, y = ecdf(data)
    plt.plot(x, y)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.xlim([0, max(x) * 1.05])
#     plt.margins(0.05)
    
plot_ecdf(app_train['OBS_60_CNT_SOCIAL_CIRCLE'], 'OBS_60_CNT_SOCIAL_CIRCLE')
plt.show()
```
